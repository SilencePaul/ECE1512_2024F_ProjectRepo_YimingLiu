```python
import torch

checkpoint = torch.load('MNIST_ConvNetD3_GraNd_exp0_epoch1_1.0_9.800000.ckpt')
print(checkpoint.keys())
```

```python
scores = checkpoint["subset"]['scores']
indices = checkpoint["subset"]['indices']
```

```python
from torchvision import datasets, transforms
data_path = "./data_path"
channel = 1
im_size = (28, 28)
num_classes = 10
mean = [0.1307]
std = [0.3081]
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=mean, std=std)])
dst_train = datasets.MNIST(data_path, train=True, download=True, transform=transform) # no augmentation
dst_test = datasets.MNIST(data_path, train=False, download=True, transform=transform)
class_names = dst_train.classes
class_map = {x:x for x in range(num_classes)}
images_all = []
labels_all = []
indices_class = [[] for c in range(num_classes)]
```

```python
images_all = [torch.unsqueeze(dst_train[i][0], dim=0) for i in range(len(dst_train))]
labels_all = [dst_train[i][1] for i in range(len(dst_train))]
```

```python
for i, lab in enumerate(labels_all):
    indices_class[lab].append(i)
```

```python
print("Number of images in the subset: ", len(indices))
print("Number of scores: ", len(scores))
print("Number of images in the dataset: ", len(dst_train))
print("Number of classes: ", num_classes)
print("Number of images per class: ", [len(indices_class[c]) for c in range(num_classes)])
```

```python
import torch

# Assuming `indices` and `scores` are lists or tensors containing the data for each label.
# Also assuming `indices_class` has been populated with the indices for each class.

sorted_tensors = []

for class_idx, class_indices in enumerate(indices_class):
    # Extract scores and indices for the current class
    class_scores = [scores[i] for i in class_indices]
    class_indices = [indices[i] for i in class_indices]
    
    # Sort the class indices based on scores in descending order
    sorted_pairs = sorted(zip(class_indices, class_scores), key=lambda x: x[1], reverse=True)
    sorted_indices = [pair[0] for pair in sorted_pairs]
    
    # Combine sorted indices and scores into a tensor
    length = len(sorted_indices)
    print("Class: ", class_idx, "Length: ", length)
    print("Sorted indices: ", sorted_indices)
    sorted_tensor = torch.tensor(sorted_indices)
    sorted_tensors.append(sorted_tensor)

torch.save(sorted_tensors, 'MNIST_ConvNet3.pt')
```
